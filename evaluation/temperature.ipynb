# Effect of temperature

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Z-scores

```python
# Load the data
data_path = 'z_scores.csv'
data = pd.read_csv(data_path)
```

```python
# Filter the entries under "experiment" that start with "temperature-" followed by a number from 0 to 1
temperature_data = data[data['experiment'].str.match(r'temperature-[0-1]')]
```

```python
# Extracting temperature values and converting them to floats for plotting
temperature_data['temperature'] = temperature_data['experiment'].str.replace('temperature-', '').astype(float)

# Plotting
plt.figure(figsize=(10, 6))

for model in temperature_data['model'].unique():
    model_data = temperature_data[temperature_data['model'] == model]
    plt.plot(model_data['temperature'], model_data['mean_z_score'], label=model, marker='o')

plt.title('Effect of Temperature on Mean Z-Scores by Model')
plt.xlabel('Temperature')
plt.ylabel('Mean Z-Score')
plt.legend()
plt.grid(True)
plt.show()
```

## Distance

```python
distance_data_path = 'mean_distances.csv'
distance_data = pd.read_csv(distance_data_path)
```

```python
# Filter the entries under "experiment" that start with "temperature-" followed by a number from 0 to 1
temperature_distance_data = distance_data[distance_data['experiment'].str.match(r'temperature-[0-1]')]

# Extracting temperature values and converting them to floats for plotting
temperature_distance_data['temperature'] = temperature_distance_data['experiment'].str.replace('temperature-', '').astype(float)
```

```python
plt.figure(figsize=(10, 6))

for model in temperature_distance_data['model'].unique():
    model_data = temperature_distance_data[temperature_distance_data['model'] == model]
    plt.plot(model_data['temperature'], model_data['mean_distances'], label=model, marker='o')

plt.title('Effect of Temperature on Mean Distances by Model')
plt.xlabel('Temperature')
plt.ylabel('Mean Distance')
plt.legend()
plt.grid(True)
plt.show()
```
